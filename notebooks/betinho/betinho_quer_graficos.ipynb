```python
import pandas as pd

import numpy as np
import matplotlib.pyplot as plt
import os
pd.set_option('display.max_columns', 93)
```

```python
df = pd.read_parquet('reduced_data/TCRF_ARCHIVE_06120018_20220523142237.parquet.parquet')
```

```python
#plot a bar graph of the data type of the df
df.dtypes.value_counts().plot(kind='bar', color=['red', 'orange', 'yellow'],title='Number of columns by data types', figsize=(12,6), rot=0,ylabel='Number of columns',xlabel='Data type')
plt.savefig('treated_data_types.png', dpi=300, bbox_inches='tight')
plt.show()
```
